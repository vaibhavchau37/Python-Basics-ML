**Q12.** Create `M = np.arange(1, 21).reshape(4, 5)`. Print the matrix, then extract:
- (a) the element at row 2, column 3
- (b) the entire second row
- (c) the entire third column
- (d) the sub-matrix consisting of the first 2 rows and last 3 columns
- (e) the last row

````python
M=np.arange(1,21).reshape(4,5)
print(M)
print('a=',M[1,2])
print('b=',M[:,3])
print('c=',M[1:4,1:4])
````

## Section 6 — Aggregations & the `axis` Argument

**Q22.** Given
```python
M = np.array([[3, 7, 2, 8],
              [5, 1, 9, 4],
              [6, 0, 3, 2]])
```
Compute:
- (a) total sum of all elements
- (b) sum of each column
- (c) sum of each row
- (d) maximum of each row
- (e) minimum of each column

````python
M=np.array([[3,7,2,8],[5,1,9,4],[6,0,3,2]])
print('sum',M.sum())
print('row sums',M.sum(axis=1))
print('col sums',M.sum(axis=0))
print('max',M.max())
print('min',M.min())
````

**Q23.** For the same matrix `M`, find the index of the maximum element in each column using `argmax`.

````python
M=np.array([[3,7,2,8],[5,1,9,4],[6,0,3,2]])
print(np.argmax(M,axis=0))
````

**Q24.** Compute the overall mean and standard deviation of `M`, rounded to 2 decimal places.

**Q27.** Generate a 1-D array of 20 random numbers drawn from a standard normal distribution (seed = 0). Then compute its mean, std, min, and max.

````python
rng=np.random.default_rng(0)
a=rng.standard_normal(20)
print(a)
print('mean',a.mean())
print('std',a.std())
````

**Q28.** Normalize the columns of a 4×3 random matrix so that each column has mean 0 and standard deviation 1. (Use broadcasting — no loops.)

````python
rng=np.random.default_rng(0)
M=rng.random((4,3))
N=(M-M.mean(axis=0))/M.std(axis=0)
print(N)
````

**Q29.** Given marks of 5 students in 3 subjects:
```python
marks = np.array([[78, 82, 90],
                  [65, 70, 72],
                  [88, 95, 92],
                  [54, 60, 58],
                  [80, 85, 87]])
```
Find: (a) each student's average, (b) each subject's average, (c) the index of the top-scoring student overall (highest total), (d) the subject in which the class performed best on average.

````python
marks=np.array([[78,82,90],[65,70,75],[88,91,85],[72,68,80],[95,94,99]])
print('student avg',marks.mean(axis=1))
print('subject avg',marks.mean(axis=0))
print('top student index',np.argmax(marks.mean(axis=1)))
````
